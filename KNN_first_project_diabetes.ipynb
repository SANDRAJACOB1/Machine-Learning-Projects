```python
#Seperating input features to x and output features to y
#x-dimension should be  2D and  y-dimension should be 1D

x=df.iloc[:,:-1]
x
#x.ndim gives the dimension
```

```python
y=df.iloc[:,-1]
y
```

```python
#Changing x as an array

x=df.iloc[:,:-1].values
x
```

```python
#Changing y as an array

y=df.iloc[:,-1].values
y
```

```python
#Separate the data into training and testing data after separating into x and y

#To split data as Training data and Testing data

from sklearn.model_selection import train_test_split
#module used is model_selection
#We have to separate both input and output into training and testing data
#creating 4 variables:
#1. training data of input
#2. testing data of input
#3. training data of output
#4. testing data of output

x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.30,random_state=1)
#there is no training size
# test_size given 0.30 because usually testing data is 30% of whole data
x_train.shape
```

```python
x_test
```

```python
# Normalization step

# Why normalization? :-
# We need to bring all the input values into a range, else: the machine will
# give output only based on the glucose values which is wrong
# 2 techniques are there;
# 1) Standard scaler
# 2) MinMax scaler
# here, we are using Standard scaler:

# 1) Standard scalar
# z=(x-u)/s
# where; z===> output
# x===> input samples
# u===> mean of training data
# s===> standard deviation of training data
# Eg: z=(6-u)/s

from sklearn.preprocessing import StandardScaler
scalar=StandardScaler() #creating object for StandardScalar
scalar.fit(x_train) #Fitting x_train, As we have to fit training data only
x_train=scalar.transform(x_train) #Coverting value of x_train to normalized
#x_train
#x_train
x_test=scalar.transform(x_test)
x_test
```

```python
#Model creation

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train) #Fitting input feature and output feature
y_pred=knn.predict(x_test) #Predicting
y_pred
```

```python
y_test #we will not get 100% accuracy
```

```python
#Performance Evaluation

#1. Accuracy Score

#Accuracy score= TP+TN/TP+TN+FP+FN
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm

#TP  #FP
#FN  #TN
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
#Accuracy score is changing during recent runtimes
#70% training and 30% testing data.
#Because the training data and testing data is splitting randomly in each run
#time. so performance changes and Accuracy score also changes in each run time.
#This is not acceptable, since we are working on the same data file and same
#code. so we define a new term  called random_state

#random_state
#Experiements shows that machine performs well when the random_state is 1,0 or
#42
#So we include random_state while spitting data to training and testing data
#When we give random_state=1 in this program, we get the maximum Accuracy score
```
